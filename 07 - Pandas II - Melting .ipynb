```python
weather = pd.read_csv("../data/weather.csv")
weather.head()
```

```python
weather_long = weather.melt(id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp')
weather_long.dropna()
```

```python
weather_tidy = (weather_long
                .pivot_table(index=['id', 'year', 'month', 'day'],
                        columns='element',
                        values='temp')
                .reset_index())
```

```python
weather_tidy.sort_values(['year', 'month', 'day'])
```

```python
weather_tidy.head()
```

```python
weather_tidy.to_csv("../data/weather_clean.csv", index=False, sep=',')
```
